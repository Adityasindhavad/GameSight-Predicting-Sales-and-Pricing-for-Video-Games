```python
# Elasticities try1 

import numpy as np
# Step 1: Identify the top 10 publishers based on global sales
avg_publisher_sales = games_df.groupby('Publisher')[['NA_revenue_mil', 'EU_revenue_mil', 'JP_revenue_mil', 'Other_revenue_mil', 'Global_revenue_mil']].mean()
avg_publisher_sales_sorted = avg_publisher_sales.sort_values(by='Global_revenue_mil', ascending=False).head(10)
top_10_publishers = avg_publisher_sales_sorted.index.tolist()

# Step 2: Filter dataset for games by these top 10 publishers
top_publisher_games = games_df[games_df['Publisher'].isin(top_10_publishers)]

# Step 3: Calculate platform-wise elasticity for each top publisher
platform_elasticities = []

# Iterate through each top publisher
for publisher in top_10_publishers:
    publisher_games = top_publisher_games[top_publisher_games['Publisher'] == publisher]
    
    # Group by game title to calculate elasticity across platforms within each game
    for game_name, group in publisher_games.groupby('Name'):
        if len(group) > 1:
            # Calculate elasticity between platforms for the same game
            for i in range(len(group) - 1):
                for j in range(i + 1, len(group)):
                    # Selecting two versions of the same game on different platforms
                    price1, price2 = group.iloc[i]['game_selling_price'], group.iloc[j]['game_selling_price']
                    sales1, sales2 = group.iloc[i]['Global_revenue_mil'], group.iloc[j]['Global_revenue_mil']
                    
                    # Calculate percentage changes
                    price_change = (price2 - price1) / price1 if price1 != 0 else np.nan
                    sales_change = (sales2 - sales1) / sales1 if sales1 != 0 else np.nan
                    
                    # Calculate elasticity if valid percentage changes exist
                    if price_change != 0 and not np.isnan(price_change) and not np.isnan(sales_change):
                        elasticity = sales_change / price_change
                        platform_elasticities.append({
                            'Publisher': publisher, 
                            'Game': game_name, 
                            'Platform 1': group.iloc[i]['Platform'],
                            'Platform 2': group.iloc[j]['Platform'],
                            'Price Change (%)': price_change * 100,
                            'Sales Change (%)': sales_change * 100,
                            'Elasticity': elasticity
                        })

# Convert results to a DataFrame for easier viewing
platform_elasticity_df = pd.DataFrame(platform_elasticities)

# Display the results for platform-wise elasticity
platform_elasticity_df.head(10)  # Displaying the top 10 rows for quick inspection
```

```python
# Step 1: Calculate platform-wise price elasticity for all games (considering all publishers)

# Initialize list to store elasticity results
platform_elasticities_all = []

# Group by game title to calculate elasticity across platforms within each game
for game_name, group in games_df.groupby('Name'):
    if len(group) > 1:
        # Calculate elasticity between platforms for the same game
        for i in range(len(group) - 1):
            for j in range(i + 1, len(group)):
                # Selecting two versions of the same game on different platforms
                price1, price2 = group.iloc[i]['game_selling_price'], group.iloc[j]['game_selling_price']
                sales1, sales2 = group.iloc[i]['Global_revenue_mil'], group.iloc[j]['Global_revenue_mil']
                publisher = group.iloc[i]['Publisher']

                # Calculate percentage changes
                price_change = (price2 - price1) / price1 if price1 != 0 else np.nan
                sales_change = (sales2 - sales1) / sales1 if sales1 != 0 else np.nan

                # Calculate elasticity if valid percentage changes exist
                if price_change != 0 and not np.isnan(price_change) and not np.isnan(sales_change):
                    elasticity = sales_change / price_change
                    platform_elasticities_all.append({
                        'Publisher': publisher,
                        'Game': game_name,
                        'Platform 1': group.iloc[i]['Platform'],
                        'Platform 2': group.iloc[j]['Platform'],
                        'Price Change (%)': price_change * 100,
                        'Sales Change (%)': sales_change * 100,
                        'Elasticity': elasticity
                    })

# Convert results to a DataFrame
platform_elasticity_df_all = pd.DataFrame(platform_elasticities_all)

# Step 2: Identifying high-potential games with high elasticity (elasticity > 1 or < -1)
high_potential_games = platform_elasticity_df_all[platform_elasticity_df_all['Elasticity'].abs() > 1]

# Step 3: Count high-potential games per publisher
publisher_potential = high_potential_games['Publisher'].value_counts()

# Displaying the results for high-potential games per publisher and a sample of the elasticity data
publisher_potential.head(10), high_potential_games.head(10)
```

The publishers with the most high-elasticity games (elasticity > 1 or < -1) include Electronic Arts, Ubisoft, and Activision. These publishers have numerous games where price changes could significantly impact sales, indicating high potential for targeted pricing adjustments.

```python
#Assumptions:

#This approach assumes that elasticity remains constant across the range of tested price changes.
#We’ll also assume that price reductions do not affect costs, allowing us to focus on maximizing sales revenue.

# Function to simulate revenue optimization for highly elastic games
def estimate_ideal_price(game_data, elasticity_threshold=1.5, price_steps=[-0.05, -0.1, -0.15, -0.2]):
    """
    Estimates the ideal price for high-elasticity games by simulating price reductions and calculating the resulting revenue.
    
    Parameters:
    - game_data: DataFrame containing game details and elasticity values.
    - elasticity_threshold: Elasticity threshold above which games are considered highly elastic.
    - price_steps: List of percentage reductions to test for optimizing revenue.
    
    Returns:
    - DataFrame with ideal price estimation for each high-elasticity game.
    """
    ideal_prices = []

    for _, row in game_data.iterrows():
        initial_price = row['game_selling_price']
        initial_sales = row['Global_revenue_mil']
        elasticity = row['Elasticity']
        
        # Only consider high elasticity games
        if abs(elasticity) > elasticity_threshold:
            max_revenue = initial_price * initial_sales  # Starting revenue
            best_price = initial_price  # Starting price

            # Simulate price reductions
            for step in price_steps:
                # Calculate new price and estimated sales
                new_price = initial_price * (1 + step)
                predicted_sales = initial_sales * (1 + (elasticity * step))
                revenue = new_price * predicted_sales

                # Check if this price point maximizes revenue
                if revenue > max_revenue:
                    max_revenue = revenue
                    best_price = new_price

            # Store result
            ideal_prices.append({
                'Game': row['Game'],
                'Publisher': row['Publisher'],
                'Initial Price': initial_price,
                'Elasticity': elasticity,
                'Optimal Price': best_price,
                'Predicted Revenue': max_revenue
            })
    
    return pd.DataFrame(ideal_prices)

# Apply this function to high-elasticity games
high_elasticity_games = high_potential_games.copy()
high_elasticity_games = high_elasticity_games.merge(games_df[['Name', 'game_selling_price', 'Global_revenue_mil']], 
                                                    left_on='Game', right_on='Name')

# Run the ideal price estimation
ideal_price_df = estimate_ideal_price(high_elasticity_games)

# Display the results
ideal_price_df.head(10)  # Showing the top 10 for overview
```

Interpretation
Price Adjustments: Games with high elasticity benefit from targeted price reductions to achieve higher predicted revenue.
Revenue Potential: Publishers can focus on games where optimal price adjustments lead to significant revenue gains, maximizing the return on high-elasticity titles.

Increased Sales Volume Compensates for Lower Price: For highly elastic games, a small reduction in price leads to a disproportionately large increase in sales. The resulting surge in volume can more than make up for the lower price, thus increasing total revenue.

Revenue Maximization in High Elasticity: When elasticity is high (absolute value > 1), the percentage increase in quantity sold is greater than the percentage decrease in price. This scenario aligns with the economic principle that lowering the price in highly elastic markets leads to higher total revenue because the gain from increased quantity outweighs the loss from a reduced price.

Price Sensitivity: High-elasticity products are price-sensitive. Therefore, reducing the price attracts more buyers or encourages existing customers to purchase more frequently, ultimately raising revenue.

# Finding Brand value of Publishers

### Make a grid for weights

```python
# Step 1: Calculate total revenue (approximate brand value based on global sales) for each publisher
publisher_revenue = games_df.groupby('Publisher')['Global_revenue_mil'].sum()

# Step 2: Calculate average critic and user ratings as a proxy for brand strength
# Assuming critic and user ratings reflect brand quality and consumer perception
publisher_ratings = games_df.groupby('Publisher')[['Critic_Score', 'User_Score']].mean()

# Step 3: Normalize and combine the data for a rough brand value metric
# To simulate "brand strength," we'll use a basic multiplier combining revenue and ratings

# Combine total revenue and average ratings into one DataFrame
publisher_brand_data = pd.DataFrame({
    'Total Revenue (millions)': publisher_revenue,
    'Avg Critic Score': publisher_ratings['Critic_Score'],
    'Avg User Score': publisher_ratings['User_Score']
})

# Normalize the revenue and scores for comparison (scaling between 0 and 1)
publisher_brand_data['Normalized Revenue'] = (publisher_brand_data['Total Revenue (millions)'] - publisher_brand_data['Total Revenue (millions)'].min()) / \
                                             (publisher_brand_data['Total Revenue (millions)'].max() - publisher_brand_data['Total Revenue (millions)'].min())
publisher_brand_data['Normalized Critic Score'] = (publisher_brand_data['Avg Critic Score'] - publisher_brand_data['Avg Critic Score'].min()) / \
                                                  (publisher_brand_data['Avg Critic Score'].max() - publisher_brand_data['Avg Critic Score'].min())
publisher_brand_data['Normalized User Score'] = (publisher_brand_data['Avg User Score'] - publisher_brand_data['Avg User Score'].min()) / \
                                                (publisher_brand_data['Avg User Score'].max() - publisher_brand_data['Avg User Score'].min())

# Calculate an approximate "Brand Value" by combining normalized revenue and ratings
# Weighing revenue more heavily as it directly impacts brand valuation # Here weights are assumed but biased towards feature importance
publisher_brand_data['Approximate Brand Value'] = (publisher_brand_data['Normalized Revenue'] * 0.6 +
                                                   publisher_brand_data['Normalized Critic Score'] * 0.2 +
                                                   publisher_brand_data['Normalized User Score'] * 0.2)

# Displaying the top 10 publishers by approximate brand value
publisher_brand_data.sort_values(by='Approximate Brand Value', ascending=False).head(10)
```

Observations
Electronic Arts leads with the highest brand value, driven by strong total revenue and favorable average ratings.
Nintendo and Activision follow, benefiting from solid revenue and high user scores.
Valve Software ranks lower in revenue but has excellent average critic and user scores, boosting its brand perception.

Interpretation
Revenue Influence: Total revenue contributes heavily to brand value, as it reflects the publisher's market presence and success.
Quality Perception: Higher critic and user scores enhance brand value, indicating consumer trust and loyalty, especially for publishers like Valve with high ratings but lower revenue.

```python
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score

# Define features and target columns
features = ['Publisher', 'Genre', 'Platform', 'game_selling_price']
targets = ['Global_revenue_mil', 'NA_revenue_mil', 'EU_revenue_mil', 'JP_revenue_mil', 'Other_revenue_mil']

# Drop rows with missing values in the relevant columns to avoid errors
cleaned_df = games_df.dropna(subset=features + targets)

# Separate features (X) and targets (y)
X = games_df[features]
y = games_df[targets]

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Adjust the preprocessor to handle unknown categories by ignoring them in OneHotEncoder
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), ['Publisher', 'Genre', 'Platform']),
        ('num', 'passthrough', ['game_selling_price'])
    ]
)

# Create the pipeline with preprocessing and regression model
model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', MultiOutputRegressor(RandomForestRegressor(n_estimators=100, random_state=0)))
])

# Train the model with the updated pipeline
model_pipeline.fit(X_train, y_train)

# Predict on the test set
y_pred = model_pipeline.predict(X_test)

# Calculate performance metrics
mae = mean_absolute_error(y_test, y_pred, multioutput='raw_values')
r2 = r2_score(y_test, y_pred, multioutput='raw_values')

# Display evaluation results
evaluation_df = pd.DataFrame({
    'Target': targets,
    'Mean Absolute Error': mae,
    'R^2 Score': r2
})

evaluation_df
```
